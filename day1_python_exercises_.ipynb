#Q2:

Take 3 numbers as input from the user and:

Store them in variables
Print the largest number without using max()

```python
numbers=[3,4,5]
large_number=numbers[0]
for number in numbers:
  if number>large_number:
    large_number=number
print(large_number)
```

#Q3:

Take a number from user and:

Print whether it is even or odd
Also print its square and cube

```python
numbers=[4,5,6,7,8,9]
dictionary_result={}
for number in numbers:
  status = "even" if number%2==0 else "odd"
  dictionary_result[number] = {"status": status, "square": number**2, "cube": number**3}
print(dictionary_result)
```

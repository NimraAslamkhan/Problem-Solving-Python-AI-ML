#Q8:

Print Fibonacci series up to n terms.

```python
def fabnochi_number(n):

  if n<=1:
     return(n)
  else:
    return fabnochi_number(n-1)+fabnochi_number(n-2)

for i in range(8):
  print(fabnochi_number(i))

```

#Q9:

Check if a number is palindrome using loop.

```python
def reurn_palandrom(n):
  return str(n)==str(n)[::-1]

print(reurn_palandrom(333333))
```

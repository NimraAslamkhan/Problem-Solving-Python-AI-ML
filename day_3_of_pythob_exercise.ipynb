#11:

 You have a dictionary representing a user profile with id, name, email, and age. You need to write a function that takes this dictionary and returns a boolean (True or False) indicating whether the profile is "complete."

A profile is defined as "complete" only if:

The age is present and is 18 or older.

Both name and email fields exist and are not empty strings.




```python
# Starting code
user_a = {'id': 1, 'name': 'Alice', 'email': 'alice@email.com', 'age': 25}
user_b = {'id': 2, 'name': 'Bob', 'email': '', 'age': 30} # Empty email
user_c = {'id': 3, 'name': 'Charlie', 'email': 'charlie@email.com'} # Missing age
user_d = {'id': 4, 'name': 'Dave', 'email': 'dave@email.com', 'age': 17} # Too young

def is_profile_complete(profile):
    if profile.get('age', 0) >= 18 and profile.get('name') and profile.get('email'):
      return True
    else:
      return False
    pass

# Test your function:
print(f"Profile A Complete: {is_profile_complete(user_a)}") # Expected: True
print(f"Profile B Complete: {is_profile_complete(user_b)}") # Expected: False
print(f"Profile C Complete: {is_profile_complete(user_c)}") # Expected: False
print(f"Profile D Complete: {is_profile_complete(user_d)}") # Expected: False
```

#3: Find Missing Number



```python
data = [1,2,3,5,6]

missing_num = []

for i in range(1, max(data)+1):
    if i not in data:
        missing_num.append(i)

print(missing_num)
```

#5: Majority Element

Find element appearing more than n/2 times

```python
data = [3,3,4,2,3,3,5]

for num in data:
  count=data.count(num)
  if count>len(data)//2:
    print(num)
    break
```

```python
nums = [2, 4, 6, 8]
target = 20

found = False

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            print(nums[i], nums[j])
            found = True

if not found:
    print("No pair found")
```

```python
nums = [1, 2, 3, 4, 5]
target = 8

seen = {}

for i, num in enumerate(nums):
    needed = target - num

    if needed in seen:
        print(seen[needed], i)
        break

    seen[num] = i
```
